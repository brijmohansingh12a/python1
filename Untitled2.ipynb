```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/TSLA.csv', low_memory=False)
```

```python
dataset.info()
```

```python


dataset.head()

```

```python
dataset.isna().sum()
```

```python
dataset_max_scaled = dataset.iloc[:, 1:]

# apply normalization techniques
for column in dataset_max_scaled.columns:
    dataset_max_scaled[column] = dataset_max_scaled[column]  / dataset_max_scaled[column].abs().max()

# view normalized data
display(dataset_max_scaled)
```

```python
x = dataset_max_scaled[['High', 'Low', 'Volume']]
y_open = dataset_max_scaled['Open']
y_close = dataset_max_scaled['Close']
```

```python
x_train, x_test = train_test_split(x)
y_open_train, y_open_test = train_test_split(y_open)
y_close_train, y_close_test = train_test_split(y_close)
```

```python
lr_open = LinearRegression()
lr_open.fit(x_train, y_open_train)
y_open_pred = lr_open.predict(x_test)

print("Coefficient (open): ", lr_open.coef_[0])
print("Intercept (open): ", lr_open.intercept_)
print("Mean Squared Error (open): ", mean_squared_error(y_open_test, y_open_pred))
print("R2_Score (open): ", r2_score(y_open_test, y_open_pred))
```

```python
lr_close = LinearRegression()
lr_close.fit(x_train, y_close_train)
y_close_pred = lr_close.predict(x_test)

print("Coefficient (close): ", lr_close.coef_[0])
print("Intercept (close): ", lr_close.intercept_)
print("Mean Squared Error (close): ", mean_squared_error(y_close_test, y_close_pred))
print("R2_Score (close): ", r2_score(y_close_test, y_close_pred))
```
